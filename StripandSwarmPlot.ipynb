```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import seaborn as sns 
import warnings
%matplotlib inline 

veri=pd.read_csv("Financial Sample.csv")
veri
```

```python
sns.stripplot(x="Year",y="Country",data=veri,jitter=True,color="purple")
```

```python
#SwarmPlot()

sns.swarmplot(x="Country",y="Year",data=veri,color="red",dodge=True)
```

```python
sns.violinplot(x="Country",y="Year",data=veri)
```

```python
sns.violinplot(x="Country",y="Year",data=veri)
sns.swarmplot(x="Country",y="Year",data=veri,color="red",dodge=True)
```
